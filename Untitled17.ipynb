```python

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Data points
x = [6000, 5000, 4000, 3000, 2000, 1000, 0, 5000, 10000, 15000, 20000, 25000, 30000]
y = [6000, 5000, 4000, 3000, 2000, 1000, 0, 5000, 10000, 15000, 20000, 25000, 30000]

# Fit a Gaussian distribution to the data points
mu, sigma = norm.fit(x)

# Generate a range of x values for the Gaussian distribution
x_range = np.linspace(min(x), max(x), 1000)

# Calculate the Gaussian distribution for the range of x values
y_gaussian = norm.pdf(x_range, mu, sigma)

# Create a scatter plot of the data points
plt.scatter(x, y)

# Plot the Gaussian distribution
plt.plot(x_range, y_gaussian, 'r', linewidth=2)

# Set labels and title
plt.xlabel('X-axis (Market Prices)')
plt.ylabel('Y-axis (Model Prices)')
plt.title('Gaussian Distribution of Market Prices vs Model Prices')

# Show the plot
plt.show()
```
